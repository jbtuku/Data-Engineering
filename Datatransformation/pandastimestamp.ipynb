```python
import pandas as pd
from google.colab import files
from datetime import datetime, timedelta


uploaded = files.upload()
filename = list(uploaded.keys())[0]
df = pd.read_csv(filename)


def convert_to_timestamp(row):
    # Extract and parse date
    date_obj = datetime.strptime(row['OPD_DATE'][:9], '%d%b%Y')
    # Create time offset from ACT_TIME
    time_delta = timedelta(seconds=int(row['ACT_TIME']))
    # Combine date and time offset
    return date_obj + time_delta


df['TIMESTAMP'] = df.apply(convert_to_timestamp, axis=1)


df.drop(columns=['OPD_DATE', 'ACT_TIME'], inplace=True)


print(" Remaining columns in the DataFrame:")
print(df.columns.tolist())


print("\n Sample data:")
print(df.head())
```
