```python
import pandas as pd
import os
from google.colab import files


uploaded = files.upload()


filename = list(uploaded.keys())[0]


df = pd.read_csv(filename)


df['HireYear'] = pd.to_datetime(df['hire_date'], errors='coerce').dt.year
invalid_hire_years = (df['HireYear'] < 2015).sum()



#  LIMIT ASSERTION 2: (NEW) Job titles must be 25 characters or less
df['title'] = df['title'].astype(str)  # ensure string type
invalid_job_titles = df['title'].apply(lambda x: len(x) > 25).sum()


print("\n=== VALIDATION REPORT ===")

print("\n Assertion 1: No employee hired before 2015")
print(f" Records violating this rule: {invalid_hire_years}")

print("\n Assertion 2: Job title must not exceed 25 characters")
print(f"Records violating this rule: {invalid_job_titles}")







```
