```python
import pandas as pd
import os
from google.colab import files


uploaded = files.upload()


filename = list(uploaded.keys())[0]


df = pd.read_csv(filename)




employee_ids = set(df['eid'].dropna())


invalid_managers = df[~df['reports_to'].isin(employee_ids) & df['reports_to'].notna()]
invalid_manager_count = len(invalid_managers)

df['HireYear'] = pd.to_datetime(df['hire_date'], errors='coerce').dt.year


hires_per_year = df['HireYear'].value_counts().sort_index()


yearly_change = hires_per_year.pct_change().abs()


violating_years = yearly_change[yearly_change > 0.75]
violating_years_count = len(violating_years)




print("Each employee must have a manager who is a known employee")
print(f" Records violating this assertion: {invalid_manager_count}")

print("\n Hires should not vary >75% year-to-year")
print(f" Number of violating years: {violating_years_count}")
print("â†’Violating years and % change:")








```
