```python
import pandas as pd
import os
from google.colab import files


uploaded = files.upload()


filename = list(uploaded.keys())[0]


df = pd.read_csv(filename)




# ----------------------------------------------------------------------------
# Summary Assertion 1: Each city must have more than one employee
# ----------------------------------------------------------------------------
employee_counts_by_city = df['city'].value_counts()
cities_with_one_employee = employee_counts_by_city[employee_counts_by_city == 1]
invalid_city_count = len(cities_with_one_employee)

print("\n=== SUMMARY ASSERTION 1 ===")
print("Assertion: Each city must have more than one employee")
print(f"Total cities violating this rule: {invalid_city_count}")
if invalid_city_count == 0:
    print("Dataset is VALID for this assertion.")
else:
    print("Dataset is NOT VALID. Cities with only one employee:")
    print(cities_with_one_employee)

# ----------------------------------------------------------------------------
#  Summary Assertion 2: Each job title must have at least 5 employees
# ----------------------------------------------------------------------------
employee_counts_by_title = df['title'].value_counts()
titles_with_few_employees = employee_counts_by_title[employee_counts_by_title < 5]
invalid_title_count = len(titles_with_few_employees)

print("\n=== SUMMARY ASSERTION 2  ===")
print("Assertion: Each job title must have at least 5 employees")
print(f"Total titles violating this rule: {invalid_title_count}")
if invalid_title_count == 0:
    print(" Dataset is VALID for this assertion.")
else:
    print(" Dataset is NOT VALID. Titles with fewer than 5 employees:")
    print(titles_with_few_employees)








```
