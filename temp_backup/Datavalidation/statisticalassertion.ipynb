```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files


uploaded = files.upload()
filename = list(uploaded.keys())[0]
df = pd.read_csv(filename)


# Assertion: The salaries are normally distributed

salary_data = df['salary'].dropna()


plt.figure(figsize=(10, 6))
sns.histplot(salary_data, kde=True, bins=30)
plt.title('Histogram of Employee Salaries')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()


mean_salary = salary_data.mean()
median_salary = salary_data.median()
std_salary = salary_data.std()
skew_salary = salary_data.skew()
kurt_salary = salary_data.kurtosis()

print("=== STATISTICAL ASSERTION 1 ===")
print("Assertion: The salaries are normally distributed")
print(f"Mean Salary: {mean_salary:.2f}")
print(f"Median Salary: {median_salary:.2f}")
print(f"Standard Deviation: {std_salary:.2f}")
print(f"Skewness: {skew_salary:.2f} (Ideal 0 for normal)")
print(f"Kurtosis: {kurt_salary:.2f} (Ideal 3 for normal)")

if abs(skew_salary) < 0.5 and abs(kurt_salary - 3) < 1:
    print("Conclusion: The salary distribution appears approximately normal.")
else:
    print("Conclusion: The salary distribution is likely NOT normal.")




df['birth_date'] = pd.to_datetime(df['birth_date'], errors='coerce')
df['age'] = (pd.to_datetime('today') - df['birth_date']).dt.days / 365.25

avg_age = df['age'].mean()

print("\n=== STATISTICAL ASSERTION 2 ===")
print("Assertion: The average employee age is between 30 and 50 years")
print(f"Average Age: {avg_age:.2f}")

if 30 <= avg_age <= 50:
    print("Conclusion: Dataset is VALID for this assertion.")
else:
    print("Conclusion: Dataset is NOT VALID for this assertion.")
```
